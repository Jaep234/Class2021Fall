```python
import numpy as np #numpy를 np로 불러온다.
import matplotlib.pyplot as plt #matplotlib 파일 속 pyplot 을 plt로 불러온다.
import IPython.display as ipd #IPython 파일 속 display 를 ipd로 불러온다.
```

```python
# parameter setting
amp = 1          # range [0.0, 1.0] #amplitute(진폭)을 1로 설정함. 
sr = 10000       # sampling rate, Hz #sampling rate(샘플링 주파수)를 1000으로 설정.
dur = 0.5        # in seconds duration을 0.5초로 설정함.
freq = 100.0     # sine frequency, Hz 주파수를 100으로 설정함.
phase = np.pi/2  # phase, radian pi/2로 값을 설정함.
```

```python
# generate time
nSamp = sr * dur #sampling rate*duration의 값을 통해 사인 그래프 내의 전체 점 개수를 설정할 수 있다.
t = np.arange(1, nSamp+1)/sr #np.arange함수를 사용하여 1부터 nSamp범위의 dot을 찍는다. arrange 함수에서 맨 끝값을 포함시키지 않기 때문에 nSamp+1을 해줌.
```

```python
#generate phase
theta=t * 2*np.pi * freq + phase #sin 함수 생성에 필수적인 theta 값 구하기. x축이 되는 시간 별 pi 값을 생성하고 phase를 통해서 곡선의 시작 위치를 정한다.
```

```python
# generate signal by cosine-phasor
s = amp * np.sin(theta) #이 과정을 통해 sin 곡선의 y축을 만든다. 진폭값을 곱해 소리의 진폭을 설정할 수 있다. 
```

```python
plt.plot(t[0:100], s[0:100], '.') #t를 x축으로, s를 y 축으로 하는 그래프를 만든다. 표식은 '.'으로
plt.xlabel('time (s)') #x 축에 'time(s)'라는 이름을 붙여줌.
plt.ylabel('real') #y 축에 'real 이라는 이름을 붙여줌.
```

```python
ipd.Audio(s, rate=sr) #ipd의Audio 함수를 이용하여 소리 추출.
```

```python
def phasor_sine(amp, sr, dur, freq, phase): #위의 과정을 하나로 통합하는 함수를 만들자.
  t = np.arange(1, sr * dur+1)/sr #시간을 설정한다.1/sr 부터 sr*dur+1/sr 의 값은 dur 범위 안에 존재.
  theta = t * 2*np.pi * freq + phase #theta 의 값을 설정한다.
  s = amp * np.sin(theta) #구한 theta의 값에 착안해 진폭을 더해준다.
  return s, t, theta #x축에 들어갈 t 값, y축데 들어갈 s 값, sin함수에 들어갈 theta 값을 출력한다.
```

```python
amp = 1; sr = 10000; dur = 1; freq = 440; phase = 0 #임의로 amp, sr, dur, freq, phase 값을 지정한다.
s, t, theta = phasor_sine(amp, sr, dur, freq, phase) #위에 정의내린 함수에 이 값을 대입하여 새로운 s, t, theta 값을 구한다.
ipd.Audio(s, rate=sr) #이를 ipd.Audio 함수에 대입한다.
```
